# Sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# z = w * x +b
def sigmoid(z):
    result = 1 / ( 1 + np.exp(-z))
    return result
```

```python
x = np.arange(-10,10,0.1)
y = sigmoid(x)
plt.plot(x,y)
plt.show()
```

```python
print(sigmoid(0))
print(sigmoid(0.1))
print(sigmoid(8))
print(sigmoid(-8))
```

# Relu

```python
# max(0,z)
def relu(z):
    if z <= 0.0:
        return 0
    else:
        return z
```

```python
x = np.arange(-10,10,0.1)
y = [relu(i) for i in x]
plt.plot(x,y)
plt.show()
```

```python
print(relu(-10))
print(relu(8))
```

# Tanh

```python
# tanh(x)= 2 / (1+e^(-2x)) = 1
def tanh(z):
    result = 2 / (1 + np.exp(-2*z)) - 1
    return result
```

```python
x = np.arange(-4,4,0.1)
y = [tanh(i) for i in x]
plt.plot(x,y)
plt.show()
```

```python
print(tanh(0))
print(tanh(-4))
print(tanh(4))
```
